```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from ast import literal_eval
```

```python
sports_df = pd.read_csv('../data/sports_10k.csv', converters={'event_list': literal_eval, 'person_list': literal_eval})
politics_df = pd.read_csv('../data/politics_10k.csv', converters={'event_list': literal_eval, 'person_list': literal_eval})
```

```python
sports_df_lda_body = pd.read_csv('../data/sports_body_10000_docs_20_topics_lda_weights.csv').join(sports_df[['date']])
sports_df_lda_title = pd.read_csv('../data/sports_title_10000_docs_20_topics_lda_weights.csv').join(sports_df[['date']])
politics_df_lda_body = pd.read_csv('../data/politics_body_10000_docs_20_topics_lda_weights.csv').join(politics_df[['date']])
politics_df_lda_title = pd.read_csv('../data/politics_title_10000_docs_20_topics_lda_weights.csv').join(politics_df[['date']])
```

```python
import utils_plot as utils
```

```python
fig, axs = plt.subplots(4, 1, sharex=True)
plt.tight_layout(pad=-0.5, w_pad=-0.5, h_pad=-0.5)

utils.plot_important_entities_over_time(sports_df,
                                             'event_list',
                                             ['olympic', 'world', 'champion', 'game', 'cup'],
                                             1,
                                             'Events',
                                             axs[0])

utils.plot_important_entities_over_time(sports_df,
                                             'person_list',
                                             ['xt', 'ferrari', 'twitter', 'nationals', 'chelsea', 'comment', 'olympic', 'sport',
                                              'spurs', 'bayern', 'buenos aires', 'grand slam', 'oguruma', 'decathon', 'ajax',
                                              'test', 'update', 'good friday'],
                                             1,
                                             'People',
                                             axs[1])

utils.plot_important_topics_over_time(sports_df_lda_body, 1, 'Topics (contents)', axs[2])
utils.plot_important_topics_over_time(sports_df_lda_title, 1, 'Topics (titles)', axs[3])

fig.set_size_inches(26, 20)
fig.savefig('../figures/insights_sports.png', format='png', dpi=100, bbox_inches='tight')
plt.show()
```
